```python
import csv
import numpy
import json
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
netFold = './Results/All/'
# traverse all the files in the folder netFold
recentData = []
for file in os.listdir(netFold):
    if file.endswith('.json'):
        with open(netFold + file) as data_file:
            data = json.load(data_file)
            recentData.append(data)
```

```python
# load the variance data into a dictionary
df2 = pd.read_csv('variance.csv')
df2
```

```python
# convert variance data into a dictionary (moviedId: variance)
dict = {}
for index, row in df2.iterrows():
    dict[str(int(row['movieId']))] = row['variance']
```

```python
dict2 = {}
for elem in recentData:
    for key in elem:
            dict2[key] = elem[key]
```

```python
recentData
```

```python
# make a new data frame using all the data from recentData
# df = pd.DataFrame(columns=['MovieID', 'Variance'])
# for i in range(len(recentData)):
#     for j in range(len(recentData[i])):
#         df.loc[len(df)] = [recentData[i][j], dict[str(recentData[i][j])]]

df = pd.DataFrame(columns=['MovieID', 'Variance', 'NAValue'])
for i in range(len(recentData)):
    for key in recentData[i]:
        df.loc[len(df)] = [key, dict[key], dict2[key]]

df['NAValue_norm'] = (df['NAValue'] - df['NAValue'].min()) / (df['NAValue'].max() - df['NAValue'].min())
df['Variance_norm'] = (df['Variance'] - df['Variance'].min()) / (df['Variance'].max() - df['Variance'].min())
df['combined'] = 0.6*df['NAValue_norm'] + df['Variance_norm']*0.4
```

```python
df.head()
```

```python
# sort the df by the combined column
df = df.sort_values(by=['combined'], ascending=False)

top_controversial = df.head(50)
top_controversial
```

```python
# store the top 50 controversial movies into a csv file
csv_file = 'top_controversial.csv'
top_controversial.to_csv(csv_file, index=False)
```
