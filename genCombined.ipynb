```python
import seaborn as sns
import cudf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
```

```python
ratingFile = './data/ratings.csv'
movieFile = './data/movies.csv'

# Read the data
ratings = cudf.read_csv(ratingFile)
movies = cudf.read_csv(movieFile)
```

```python
print(ratings.head())
```

```python
print(movies.head())
```

```python
possibleRatings = ratings['rating'].unique()
```

```python
# create a blnk dataframe
df = cudf.DataFrame(columns=['movieId', 'title', 'genre', '0.5', '1.0', '1.5', '2.0', '2.5', '3.0', '3.5', '4.0', '4.5', '5.0'])
df
```

```python
# fill the dataframe

df['movieId'] = movies['movieId']
df['title'] = movies['title']
df['genre'] = movies['genres']
df
```

```python
for ind in df.index:
    df.loc[ind, '0.5'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 0.5)].shape[0]
    df.loc[ind, '1.0'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 1.0)].shape[0]
    df.loc[ind, '1.5'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 1.5)].shape[0]
    df.loc[ind, '2.0'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 2.0)].shape[0]
    df.loc[ind, '2.5'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 2.5)].shape[0]
    df.loc[ind, '3.0'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 3.0)].shape[0]
    df.loc[ind, '3.5'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 3.5)].shape[0]
    df.loc[ind, '4.0'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 4.0)].shape[0]
    df.loc[ind, '4.5'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 4.5)].shape[0]
    df.loc[ind, '5.0'] = ratings[(ratings['movieId'] == df.loc[ind, 'movieId']) & (ratings['rating'] == 5.0)].shape[0]
    if ind % 100 == 0:
        print("Number of Movies Processed: ", ind)
```

```python
# store the dataframe in a new csv file
df.to_csv('combined.csv')
```
