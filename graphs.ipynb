```python
new_ratings = './Data/ml-latest/newRatings.csv'

ratings = pd.read_csv(new_ratings)
ratings = ratings.sort_values(by=['movieId'])
ratings = ratings.drop('timestamp', axis=1)
ratings.head()
```

```python
# make a new dictionary with movieId as key and have another dictionary with userId as key and rating as value
# this is to make it easier to find the rating for a specific movie and user

movie_ratings = {}
for index, row in ratings.iterrows():
    if row['movieId'] not in movie_ratings:
        movie_ratings[row['movieId']] = {}
    movie_ratings[row['movieId']][row['userId']] = row['rating']
```
